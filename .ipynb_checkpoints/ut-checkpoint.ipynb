```python
from lightgbm import LGBMRegressor
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
```

```python
df = df.sort_values(["asset", "timestamp"]).reset_index(drop=True)
```

```python
for tgt in ["vwap", "e", "h", "x"]:
    df[f"{tgt}_next"] = df.groupby("asset")[tgt].shift(-1)
```

```python
df = df.dropna(subset=["vwap_next", "e_next", "h_next", "x_next"]).reset_index(drop=True)
```

```python
lag_list = [1, 2, 3]

for col in ["vwap", "e", "h", "x", "sigma", "spread", "spreadper"]:
    for lag in lag_list:
        df[f"{col}_lag{lag}"] = df.groupby("asset")[col].shift(lag)

# Remove rows with NA from lag creation
df = df.dropna().reset_index(drop=True)
```

```python
le_asset = LabelEncoder()
le_gap = LabelEncoder()

df["asset_enc"] = le_asset.fit_transform(df["asset"])
df["gap_enc"]   = le_gap.fit_transform(df["gap"])
```

```python
df.tail()
```

```python
feature_cols = [
    'klineacc','spread','spreadper','x','vwap','deviation','ratio',
    'term','sigma','e','h','gaplimit',
    'asset_enc', 'gap_enc'
] + [
    c for c in df.columns
    if "lag" in c
]

target_col = "vwap_next"
```

```python
X = df[feature_cols].astype(float)
y = df[target_col].astype(float)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.1, shuffle=True, random_state=42
)

model = LGBMRegressor(
    n_estimators=600,
    learning_rate=0.03,
    max_depth=-1,
    subsample=0.9,
    colsample_bytree=0.8
)

model.fit(X_train, y_train)


print("\nModel training completed.")
print("Validation R²:", model.score(X_test, y_test))
print("Sample prediction:", model.predict(X_test.iloc[:5]))
```

```python
combined = pd.concat([df, ndf], ignore_index=True)

combined = combined.sort_values(["asset", "timestamp"]).reset_index(drop=True)

# encode ndf asset/gap using the same encoders from training
combined["asset_enc"] = le_asset.transform(combined["asset"])
combined["gap_enc"]   = le_gap.transform(combined["gap"])

# generate lag features on combined data
for col in ["vwap", "e", "h", "x", "sigma", "spread", "spreadper"]:
    for lag in lag_list:
        combined[f"{col}_lag{lag}"] = combined.groupby("asset")[col].shift(lag)

# extract ONLY the latest row per asset from ndf (prediction point)
ndf_pred = (
    combined[combined["timestamp"].isin(ndf["timestamp"])]
    .groupby("asset")
    .tail(1)
)

# keep only rows with valid lag data
ndf_pred = ndf_pred.dropna(subset=[f"vwap_lag{lag_list[-1]}"])

###############################################################################
# 10) RUN PREDICTIONS
###############################################################################

X_ndf = ndf_pred[feature_cols].astype(float)

ndf_pred["prediction"] = model.predict(X_ndf)

print("\n=== PREDICTIONS FOR NEW DATA (NDF) ===")
print(ndf_pred[["asset", "timestamp", "prediction"]])
```
