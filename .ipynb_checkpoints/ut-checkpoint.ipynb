```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import TimeSeriesSplit
from xgboost import XGBRegressor
import pickle

# --------------------------
# 1. Load your data
# --------------------------
# Example: replace with your CSV or DataFrame
# data = pd.read_csv('your_data.csv')

# Check columns exist
required_cols = ['klineacc','spread','spreadper','x','vwap','deviation','ratio','term','sigma','e','h','asset','timestamp','gap']
for c in required_cols:
    if c not in data.columns:
        raise ValueError(f"Missing column: {c}")

# --------------------------
# 2. Define features
# --------------------------
feature_cols = ['x','h','e','ratio','term','sigma','deviation','spread','spreadper','klineacc']

# --------------------------
# 3. Prepare scalers per asset
# --------------------------
assets = data['asset'].unique()
scalers = {}
for asset in assets:
    scalers[asset] = StandardScaler()
    idx = data['asset'] == asset
    data.loc[idx, feature_cols] = scalers[asset].fit_transform(data.loc[idx, feature_cols])

# --------------------------
# 4. Train models per asset and gap
# --------------------------
gap_types = ['1m','3m','5m','15m','30m','1h','2h','4h','6h','8h','12h']
models = {asset:{} for asset in assets}

min_samples = 10  # only train if enough rows

for asset in assets:
    for gap in gap_types:
        gap_data = data[(data['asset']==asset) & (data['gap']==gap)]
        if len(gap_data) < min_samples:
            # print(f"Skipping training for {asset} {gap} (not enough rows)")
            continue
        
        X_train = gap_data[feature_cols]
        y_train = gap_data['vwap'].shift(-1)  # predict next vwap
        y_train = y_train.fillna(method='ffill')  # handle last NaN

        model = XGBRegressor(n_estimators=300, max_depth=5, learning_rate=0.05)
        model.fit(X_train, y_train)
        models[asset][gap] = model

# Optional: save models for later
# with open('models.pkl','wb') as f:
#     pickle.dump(models,f)

# --------------------------
# 5. Make predictions
# --------------------------
predictions = []

for asset in assets:
    for gap in gap_types:
        if gap not in models[asset]:
            continue  # skip untrained
        
        gap_data = data[(data['asset']==asset) & (data['gap']==gap)].copy()
        gap_data[feature_cols] = scalers[asset].transform(gap_data[feature_cols])
        gap_data['pred_next_vwap'] = models[asset][gap].predict(gap_data[feature_cols])
        predictions.append(gap_data[['timestamp','asset','gap','pred_next_vwap']])

pred_df = pd.concat(predictions).reset_index(drop=True)

# --------------------------
# 6. Output
# --------------------------
pred_df.to_csv('predicted_vwap.csv', index=False)
print("Predictions saved to predicted_vwap.csv")
```

```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestRegressor

df = df.sort_values(["asset", "gap", "timestamp"]).reset_index(drop=True)

df["true_next_vwap"] = df.groupby(["asset","gap"])["vwap"].shift(-1)

df = df.dropna(subset=["true_next_vwap"])

FEATURES = [
    "klineacc","spread","spreadper","x","vwap","deviation","ratio",
    "term","sigma","e","h","gaplimit"
]

TARGET = "true_next_vwap"

for f in FEATURES:
    if f not in df.columns:
        raise Exception(f"Feature missing: {f}")

# Clean NaN rows
df = df.dropna(subset=FEATURES + [TARGET])

X = df[FEATURES]
y = df[TARGET]

model = RandomForestRegressor(
    n_estimators=500,
    max_depth=12,
    random_state=42,
    n_jobs=-1
)

model.fit(X, y)

pred_rows = df.sort_values(["asset","gap","timestamp"]).groupby(
    ["asset","gap"]
).tail(1)

pred_X = pred_rows[FEATURES]
pred_y = model.predict(pred_X)

pred_rows["pred_next_vwap"] = pred_y

pred_rows[[
    "asset","gap","timestamp","vwap","true_next_vwap","pred_next_vwap"
]].to_csv("final_predictions.csv", index=False)

print("Prediction complete. One row per assetgap delivered.")
```

```python
ddf = pd.read_csv("final_predictions.csv")

# ----------------------
# 2. Calculate row-wise errors
# ----------------------
ddf['error'] = ddf['pred_next_vwap'] - ddf['true_next_vwap']
ddf['abs_error'] = ddf['error'].abs()
ddf['squared_error'] = ddf['error'] ** 2
ddf['direction_correct'] = np.sign(ddf['pred_next_vwap'] - ddf['vwap']) == np.sign(ddf['true_next_vwap'] - ddf['vwap'])

# ----------------------
# 3. Per-asset aggregation
# ----------------------
agg_asset = ddf.groupby('asset').agg(
    RMSE=('squared_error', lambda x: np.sqrt(x.mean())),
    MAE=('abs_error', 'mean'),
    Direction_Acc=('direction_correct', 'mean'),
    Count=('asset', 'count')
).reset_index()

# ----------------------
# 4. Per-asset, per-gap aggregation (optional)
# ----------------------
agg_asset_gap = ddf.groupby(['asset', 'gap']).agg(
    RMSE=('squared_error', lambda x: np.sqrt(x.mean())),
    MAE=('abs_error', 'mean'),
    Direction_Acc=('direction_correct', 'mean'),
    Count=('asset', 'count')
).reset_index()

# ----------------------
# 5. Overall metrics
# ----------------------
overall_rmse = np.sqrt((ddf['squared_error'].sum()) / len(ddf))
overall_mae  = ddf['abs_error'].mean()
overall_acc  = ddf['direction_correct'].mean()

# ----------------------
# 6. Save results
# ----------------------
agg_asset.to_csv("eval_per_asset.csv", index=False)
agg_asset_gap.to_csv("eval_per_asset_gap.csv", index=False)
ddf.to_csv("eval_per_row.csv", index=False)

# ----------------------
# 7. Print summary
# ----------------------
print(f"Overall RMSE: {overall_rmse:.6f}, MAE: {overall_mae:.6f}, Direction Acc: {overall_acc:.4f}")
print("\nPer-Asset Summary:")
print(agg_asset)

print(f"Overall RMSE: {overall_rmse:.6f}, MAE: {overall_mae:.6f}, Direction Acc: {overall_acc:.4f}")
print("\nPer-Asset-GAP Summary:")
print(agg_asset_gap)
```
